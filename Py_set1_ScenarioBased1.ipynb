Scenario 05: A text-processing tool reverses a given word or sentence for formatting purposes.  Write logic to take a word or sentence as input and produce its reversed version.


```python
# Take input from the user
text = input("Enter a word or sentence: ")

# Reverse the text
reversed_text = ""
for i in range(len(text) - 1, -1, -1):
    reversed_text += text[i]

# Print the reversed result
print("Reversed text:", reversed_text)

```

Scenario 07: A retail store offers a 20% discount if a customer’s total order exceeds $100. Write logic to calculate the final amount to be paid after applying the discount.


```python
#Total Order Value Input
Value=int(input("Total Order Value of the Customer:"))

if Value>100:
    print(f"Customer is eligible for 20% discount. Final Amount to be paid is {Value-(Value*0.2)}")
else:
    print("Customer is not eligible for discount")
```

Scenario 09: A calendar system verifies whether a given year is a leap year based on standard leap year rules.  Write logic to determine whether a given year is a leap year.

```python
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not a leap year")
```
