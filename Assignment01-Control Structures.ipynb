```python
#Body Mass Index (BMI)
BMI=float(input("Enter the BMI Index:"))

if (BMI<18.5):
    print("Under Weight")
elif (BMI<=24.9):
    print("Normal Range")
elif (BMI<=29.9):
    print("Over weight")
elif (BMI<=34.9):
    print("Obese class I") 
elif (BMI<=39.9):
    print("Obese class II") 
elif (BMI>=40):
    print("Obese class III") 
```
