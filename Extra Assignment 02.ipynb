```python
#Print Odd numbers in the List
TupA=(20,10,16,19,25,1,276,188)

for Num in TupA:
    if (Num%2==1):
        print(Num, "is Odd") 
        
```

```python
#Print Even Numbers
TupA=(20,10,16,19,25,1,276,188)

for Num in TupA:
    if (Num%2==0):
        print(Num, "is Even") 
```
